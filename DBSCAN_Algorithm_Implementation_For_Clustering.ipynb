```python
#This is a function for implementaiton of the algorithm DBSCAN. It will identify a point belonging to a cluster given minimum distance
# and number of points to be considered. Output will give the label for clusters (1,2,3..) and -1 will indicate the point
# is a noise point.

import numpy

# Main funciton for DBSCAN clustering
def DBSCAN(data, eps, minPts):
    n = len(data)
    labels = [0]*n   # make a list of zero's and iniliaze every points with it.
    C = 0            # c= 0 means point hasn't been considered yet. Intialize all point with zero
    for point in range(0, n):
        if not (labels[point] == 0):  #i fpoint has already been considered and labeled , continue.
            continue
        
        neighbor = rangeQuery(data, point, eps)  #function to find all the neighbors based on given eps 
        if len(neighbor) < minPts:
            labels[point] = -1        #label = -1 indicate point is a noise point
        else: 
            C += 1
            expandCluster(data, labels, point, neighbor, C, eps, minPts)
    
    return labels

# Function to expand clusters
def expandCluster(data, labels, point, neighbor, C, eps, minPts):
   
    labels[point] = C
    i = 0
    while i < len(neighbor):       
        newPoint = neighbor[i]
        if labels[newPoint] == -1:
            labels[newPoint] = C

        elif labels[newPoint] == 0:
            labels[newPoint] = C
            newNeighborPoints = rangeQuery(data, newPoint, eps)
    
            if len(newNeighborPoints) >= minPts:
                neighbor = neighbor + newNeighborPoints
        i += 1        

#Function to look for neighbors of a given point based on given eps        
def rangeQuery(data, point, eps):
    neighbors = []
    n = len(data)
    # For each point in the dataset...
    for newPoint in range(0, n):
        
        # If the distance is below the threshold, add it to the neighbors list.
        if numpy.linalg.norm(data[point] - data[newPoint]) < eps:
            neighbors.append(newPoint)
            
    return neighbors
```
